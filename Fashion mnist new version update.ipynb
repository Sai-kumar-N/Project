```python
X_valid, X_train = X_train_full[:5000], X_train_full[5000:] 
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
"Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(True)
    plt.imshow(X_train[i],cmap='gray')
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

```python
from tensorflow.keras import Sequential 
```

```python
from tensorflow.keras.layers import Dense, Flatten
```

```python
model = Sequential()
model.add(Flatten(input_shape=[28, 28]))
model.add(Dense(100, activation="relu"))
model.add(Dense(300, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",
optimizer="adam",
metrics=["accuracy"])
```

```python
m = model.fit(X_train, y_train, epochs=20,validation_data=(X_valid, y_valid))
```

```python
import pandas as pd
pd.DataFrame(m.history).plot(figsize=(8, 5))
plt.grid(True)
#plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.show()
```

```python
plt.plot(m.history['accuracy'], label = 'acc')
plt.plot(m.history['val_accuracy'], label = 'val_acc')
plt.legend()
```

```python
plt.plot(m.history['loss'], label = 'loss')
plt.plot(m.history['val_loss'], label = 'val_loss')
plt.legend()
```

```python
model.evaluate(X_test, y_test)
```

```python
X_new = X_test[:10]
y_proba = model.predict(X_new)
y_proba.round(2)
```

```python
y_pred = model.predict_classes(X_new)
y_pred
```
